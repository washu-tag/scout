# Scout RADS Dashboard

Interactive dashboard for analyzing LI-RADS (Liver Imaging Reporting and Data System) scores from radiology reports.

Features:
- **Score Extraction**: Automatically extract and parse LI-RADS scores from report text
- **Population Statistics**: Age, sex, race demographics and overall statistics
- **Score Distribution**: Visual breakdown of RADS scores across the population
- **Time Comparisons**: Compare score distributions across different time periods (month-to-month, quarter-to-quarter, year-over-year)
- **Patient Progression**: Longitudinal analysis tracking individual patients' score changes over time
- **Demographics Breakdown**: Score distribution by age groups, sex, and race
- **Report Browser**: Navigate and review individual reports with highlighted RADS mentions
- **Export**: Download results as CSV for further analysis
